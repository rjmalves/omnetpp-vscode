```python
import pandas as pd
SCALARS_FILE = "./scalars.csv"
VECTORS_FILE = "./vectors.csv"
```

```python
df_scalars = pd.read_csv(SCALARS_FILE)
df_scalars
```

```python
df_vectors = pd.read_csv(VECTORS_FILE)
df_vectors
```

```python
src_module = "WirelessVSCodeNetwork.host1.app[0]"
dst_module = "WirelessVSCodeNetwork.host3.app[0]"
src_attr = "packetSent:count"
dst_attr = "packetReceived:count"
sent = df_scalars.loc[(df_scalars["module"] == src_module) & (df_scalars["name"] == src_attr) & (df_scalars["attrname"].isna()), "value"]
recv = df_scalars.loc[(df_scalars["module"] == dst_module) & (df_scalars["name"] == dst_attr) & (df_scalars["attrname"].isna()), "value"]
per = 100 * (1.0 - (float(recv) / float(sent)))
print(f"Packet Error Rate = {per}")
```

```python
dst_module = "WirelessVSCodeNetwork.host3.app[0]"
throughput_name = "throughput:vector"
recv_data = df_vectors.loc[(df_vectors["type"] == "vector") & (df_vectors["module"] == dst_module), :]
throughput_data = recv_data.loc[recv_data["name"] == throughput_name, ["vectime", "vecvalue"]]
throughput_data
```

```python
x = [float(t) for t in throughput_data["vectime"].tolist()[0].split(" ")]
y = [float(t) for t in throughput_data["vecvalue"].tolist()[0].split(" ")]
plot_df = pd.DataFrame(data={"time": x, "throughput": y})
plot_df
```

```python
plot_df.plot(x="time", y="throughput")
```
